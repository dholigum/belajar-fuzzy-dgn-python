### Studi Kasus: Fungsi Keanggotaan Fuzzy

```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### SOAL 1

Robot pendulum terbalik sedang didesain menggunakan logika fuzzy untuk menjaganya agar tidak jatuh. Buatlah menjadi 5 kategori kemiringan (tentukan sendiri) dengan menggunakan fungsi keanggotaan gaussian jika :
0 derajat = tegak
lebih dari 45 derajat = pasti jatuh

Rule:
1. Tegak: 0 derajat
2. Agak Miring: 12,5 derajat
3. Miring: 25 derajat
4. Hampir Jatuh: 37,5 derajat
5. Pasti Jatuh: 45 derajat

```python
x = np.arange(0, 45, 0.1)

tegak = fuzz.gaussmf(x, 0, 4)
agak_miring = fuzz.gaussmf(x, 12.5, 4)
miring = fuzz.gaussmf(x, 25, 4)
hampir_jatuh = fuzz.gaussmf(x, 37.5, 4)
pasti_jatuh = fuzz.gaussmf(x, 45, 4)

plt.figure()
plt.plot(x, tegak, 'b', linewidth=1.5, label="Tegak")
plt.plot(x, agak_miring, 'k', linewidth=1.5, label="Agak Miring")
plt.plot(x, miring, 'm', linewidth=1.5, label="Miring")
plt.plot(x, hampir_jatuh, 'r', linewidth=1.5, label="Hampir Jatuh")
plt.plot(x, pasti_jatuh, 'c', linewidth=1.5, label="Pasti Jatuh")
plt.title('Fuzzy Robot Pendulum Terbalik')
plt.ylabel('Keanggotaan')
plt.xlabel('Kemiringan (derajat)')
plt.legend(loc="center right", bbox_to_anchor=(1.25, 0.5), ncol=1, fancybox=True, shadow=True)
```

### SOAL 2

Robo boat sedang akan mengikuti lomba balap tingkat nasional. namun perlu diatur kecepatannya agar dapat bermanuver dengan baik. Buatlah menjadi 5 kategori kecepatan (tentukan sendiri) dengan menggunakan fungsi keanggotaan trapesium jika :
lebih dari 0 km/jam = bergerak
60 km/jam = kecepatan maksimal

Rule:

1. Mulai Bergerak: [0 5 10 15]
2. Kec. Lambat: [10 15 20 25]
3. Kec. Sedang: [20 25 30 35]
4. Kec. Hampir maksimal: [30 35 45 50]
5. Kec. Maksimal: [45 50 55 60]

```python
x = np.arange(0, 60, 0.1)

mulai_gerak = fuzz.trapmf(x, [0, 5, 10, 15])
kec_lambat = fuzz.trapmf(x, [10, 15, 20, 25])
kec_sedang = fuzz.trapmf(x, [20, 25, 35, 40])
kec_hampir_maks = fuzz.trapmf(x, [35, 40, 45, 50])
kec_maks = fuzz.trapmf(x, [45, 50, 55, 60])

plt.figure()
plt.plot(x, mulai_gerak, 'b', linewidth=1.5, label="Mulai Bergerak")
plt.plot(x, kec_lambat, 'k', linewidth=1.5, label="Kec. Lambat")
plt.plot(x, kec_sedang, 'm', linewidth=1.5, label="Kec. Sedang")
plt.plot(x, kec_hampir_maks, 'r', linewidth=1.5, label="Kec. Hampir maksimal")
plt.plot(x, kec_maks, 'c', linewidth=1.5, label="Kec. Maksimal")
plt.title('Fuzzy Robot Boat')
plt.ylabel('Keanggotaan')
plt.xlabel('Kecepatan (km/jam)')
plt.legend(loc="center right", bbox_to_anchor=(1.25, 0.5), ncol=1, fancybox=True, shadow=True)
```

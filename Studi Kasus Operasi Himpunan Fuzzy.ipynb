### Studi Kasus: Operasi Himpunan Fuzzy

```python
import skfuzzy as sk
import numpy as np
import matplotlib.pyplot as plt
```

### SOAL 1

Himpunan A => himpunan fuzzy berbentuk fungsi segitiga dengan nilai anggota dr 0 - 10 di mana ketika 0 derajat keanggotaannya 0, ketika 7 derajat keanggotaannya 1, dan ketika 10 derajat keanggotannya 0

```python
x_qual = np.arange(0, 11, 1)
qual_lo = sk.trimf(x_qual, [0, 7, 10])

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

Himpunan B => himpunan fuzzy berbentuk fungsi trapesium dengan nilai anggota dr 0 - 10 di mana ketika 0 derajat keanggotaannya 0, ketika 4 derajat keanggotaannya 1, ketika 7 derajat keanggotaannya 1, dan ketika 10 derajat keanggotannya 0.

```python
x_qual = np.arange(0, 11, 1)
qual_lo = sk.trapmf(x_qual, [0, 4, 7, 10])

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

A irisan B = ?. Gambarkan grafiknya

```python
x_qual = np.arange(0, 11, 1)
qual_A = sk.trimf(x_qual, [0, 7, 10])
qual_B = sk.trapmf(x_qual, [0, 4, 7, 10])

irisan = sk.fuzzy_and(x_qual, qual_A, x_qual, qual_B)

print('A irisan B = \n', irisan)

fig, ax = plt.subplots()
ax.plot(qual_A)
ax.set_xlabel('Fungsi A', fontsize = 12)
ax.set_ylabel('Derajat Keanggotaan', fontsize = 12)

fig2, ax2 = plt.subplots()
ax2.plot(qual_B)
ax2.set_xlabel('Fungsi B', fontsize = 12)
ax2.set_ylabel('Derajat Keanggotaan', fontsize = 12)

fig3, ax3 = plt.subplots()
ax3.plot(irisan[1])
ax3.set_xlabel('Fungsi (A irisan B)', fontsize = 12)
ax3.set_ylabel('Derajat Keanggotaan', fontsize = 12)
```

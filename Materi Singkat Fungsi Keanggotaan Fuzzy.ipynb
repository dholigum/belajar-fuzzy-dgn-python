### Materi: Fungsi Keanggotaan Fuzzy

Fungsi Keanggotaan dalam Fuzzy merupakan bagian penentu tingkat atau derajat keanggotaan dari tiap elemen dalam semesta dari sebuah Himpunan. Terdapat beberapa jenis fungsi keanggotaan antara lainnya:
1. Fungsi Kurva Segitiga
2. Fungsi Kurva Trapesium
3. Fungsi Kurva Lonceng (Bell)
4. Fungsi Kurva Gaussian
5. Fungsi Kurva Sigmoid
6. Fungsi Kurva S
7. Fungsi Kurva Z

```python
import skfuzzy as sk
import numpy as np
import matplotlib.pyplot as plt
```

### Fungsi Kurva Segitiga

<img src="https://static-01.hindawi.com/articles/tswj/volume-2014/347619/figures/347619.fig.001.jpg">

```python
x_qual = np.arange(0, 11, 1)

# trimf(array_range, nilai_awal_kurva, nilai_tengah, nilai_akhir_kurva)
qual_lo = sk.trimf(x_qual, [0, 7, 10])

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva Trapesium

<img src="https://www.gaetanointrieri.com/wp-content/uploads/2015/10/Senza-titolo-2.png">

```python
x_qual = np.arange(0, 11, 1)

# trapmf(array_range, nilai_awal_kurva, nilai_konstant_atas,
#             nilai_akhir_kurva)
qual_lo = sk.trapmf(x_qual, [0, 4, 7, 10])

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva Lonceng (Bell)

<img src="https://d3i71xaburhd42.cloudfront.net/9e2170c0962929d7766d69b0f21e2e8e647ede7a/13-Figure10-1.png">

```python
x_qual = np.arange(0, 11, 1)

# mirip seperti fungsi gaussian, tetapi puncaknya bisa agak konstant
# gaussmf(array_range, nilai_lebar_kurva, nilai_kelengkungan_atas,
#             nilai_tengah_kurva)
qual_lo = sk.gbellmf(x_qual, 3, 3, 5)

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva Gaussian

<img src="https://www.researchgate.net/profile/Christian_Vasseur/publication/220601674/figure/fig10/AS:364220957904902@1463848451748/Gaussian-membership-function-for-8-x-I-I.png">

```python
x_qual = np.arange(0, 11, 1)

# mirip seperti fungsi bel, tetapi puncaknya lancip
# gaussmf(array_range, nilai_tengah, nilai_lebar_kurva_yk)
qual_lo = sk.gaussmf(x_qual, 5, 2)

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva Sigmoid

<img src="https://www.researchgate.net/profile/Julie_Wall/publication/234128569/figure/fig1/AS:299937389662208@1448522054369/Sigmoid-Membership-Function.png">

```python
x_qual = np.arange(0, 11, 1)

# sigmf(array_range, nilai_tengah, nilai_lebar_kurva)
qual_lo = sk.sigmf(x_qual, 5, 2)

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva S

<img src="https://d3i71xaburhd42.cloudfront.net/9e2170c0962929d7766d69b0f21e2e8e647ede7a/8-Figure3-1.png">

```python
x_qual = np.arange(0, 11, 1)

# fungsi S tidak bisa diatur lebarnya seperti fungsi kurva sigmoid
# sigmf(array_range, nilai_awal_lalu_naik, nilai_akhir)
qual_lo = sk.smf(x_qual, 0, 10)

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```

### Fungsi Kurva Z

```python
x_qual = np.arange(0, 11, 1)

# kebalikan dari fungsi S
# sigmf(array_range, nilai_awal_lalu_turun, nilai_akhir)
qual_lo = sk.zmf(x_qual, 0, 10)

print(qual_lo)

fig,ax = plt.subplots()
ax.plot(qual_lo)
ax.set_ylabel('Derajat Keanggotaan', fontsize=12)
```
